```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as BS
import re
```

```python
#scraping in postal codes
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; rv:20.0) Gecko/20100101 Firefox/20.0'}
url = "https://www.scouting.org/resources/los/states/"
response = requests.get(url, headers=headers)
soup= BS(response.text)
abbreviations = pd.read_html(response.text)[0]
abbreviations

#scraping in district number
url = 'https://www.britannica.com/topic/United-States-House-of-Representatives-Seats-by-State-1787120'
response = requests.get(url)
soup= BS(response.text)
district_number = pd.read_html(response.text)[0]
district_number

#pd.merge(abbreviations, district_number, left_on=['State'], right_on=['state'], how='left')
state_postal_rep = abbreviations.join(district_number.set_index(['state']), on=['State'], how='left').dropna()
state_postal_rep['representatives'] = state_postal_rep['representatives'].astype('int')
```

```python
#create dictionary postal codes:reps
state_rep_dict = dict(zip(state_postal_rep.Postal, state_postal_rep.representatives))
```

```python
table= []  

def get_district(state, district_number):
    global table  
    zfill = str(district_number).zfill(2)
    base_url = 'https://www.opensecrets.org/races/summary?cycle=2020&id={}{}&spec=N'.format(state, zfill)
    response = requests.get(base_url)
    soup = BS(response.text)
    candidate = pd.read_html(str(soup.find('table', attrs={'class' : 'DataTable'})))[0]
    candidate["District"] = zfill
    candidate['State'] = state
    table.append(candidate)
    return table
```

```python
pd.options.display.max_rows=1000
for state, reps in state_rep_dict.items():
    for district in range (1,reps+1):
        get_district(state,district)
table_concat = pd.concat(table, ignore_index=True)   
table_concat  
```
